```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Sales.xlsx")
df.boxplot(column="Sale_amt",by="Region",patch_artist=True, vert=False)
plt.xlabel("Sale_amt",color="g")
plt.ylabel("Region",color="g")
plt.title("Boxplot of Sale_amt by Region",color="r")
plt.show()
```

##Histogram/displot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Sales.xlsx")
sns.displot(df["Sale_amt"])
plt.title("Histogram of Sale_amt",color="r")
plt.xlabel("Sale_amt",color="g")
plt.ylabel("Frequency",color="g")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Sales.xlsx")
sns.displot(df["Sale_amt"],kde=True)
plt.xlabel("Sale_amt",color="g")
plt.ylabel("Frequency",color="g")
plt.title("Histogram of Sale_amt",color="r")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_excel("Sales.xlsx")
sns.displot(df["Sale_amt"],kde=True,rug=True,color="g")
plt.xlabel("Sale_amt",color="g")
plt.ylabel("Frequency",color="g")
plt.title("Histogram of Sale_amt",color="r")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Sales.xlsx")
df.plot(x="OrderDate",y="Sale_amt",marker="o",color="r",markeredgecolor="cyan")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("Sales.xlsx")
saledater=df.groupby("OrderDate")["Sale_amt"].sum()
saledater.plot(kind="line",marker="x")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.close()

df = pd.read_excel("Sales.xlsx")

plt.figure(figsize=(14,5))

plt.subplot(1,2,1)
sns.boxplot(x="Region", y="Sale_amt", data=df)
plt.title("Sales by Region")

plt.subplot(1,2,2)
sns.boxplot(x="Item", y="Sale_amt", data=df)
plt.title("Sales by Item")
plt.xticks(rotation=45)   # rotate x-axis labels for better visibility

plt.tight_layout()
plt.show()
```
