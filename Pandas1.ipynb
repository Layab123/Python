#Exception Handling

```python
try:
  print(a)
except:
  print("An exception occured")
```

```python
try:
  print*(2 *7)
except:
  print("An exception occured")

```

```python
try:
  print(a)
except NameError:
  print("Variable a is not defined")
except:
  print("Something else went wrong")
```

```python
try:
  print(x)
except:
  print("Something went wrong")
finally:
  print("The 'try except' is finished")
```
